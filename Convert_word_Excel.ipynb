```python
import pandas as pd
from docx import Document
from google.colab import files

# === Step 1: Upload file ===
print("📂 Please upload your Word (.docx) file")
uploaded = files.upload()
docx_file = list(uploaded.keys())[0]

# === Step 2: Load document ===
doc = Document(docx_file)

# === Step 3: Extract tables ===
all_tables = []

for i, table in enumerate(doc.tables, start=1):
    data = []
    for row in table.rows:
        data.append([cell.text.strip() for cell in row.cells])
    df = pd.DataFrame(data)
    all_tables.append((f"Table_{i}", df))

# ✅ CHECK FIRST — ONLY CONTINUE IF TABLES EXIST
if not all_tables:
    print("❌ No tables found in this Word document.")
else:
    # === Step 4: Ask user preference ===
    choice = input(
        "Do you want all tables merged into ONE sheet (type 'merge') "
        "or each table in a SEPARATE sheet (type 'separate')? "
    ).strip().lower()

    # === Step 5: Process ===
    if choice == "merge":
        merged_df = pd.concat([df for _, df in all_tables], ignore_index=True)

        print("\n🔎 Preview of merged data:")
        display(merged_df)

        output_file = "Word_to_Excel_Merged.xlsx"
        merged_df.to_excel(output_file, index=False)

    elif choice == "separate":
        output_file = "Word_to_Excel_SeparateSheets.xlsx"

        with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
            for sheet_name, df in all_tables:
                df.to_excel(writer, index=False, sheet_name=sheet_name)

        print(f"\n🔎 Preview of {all_tables[0][0]}:")
        display(all_tables[0][1])

    else:
        print("❌ Invalid choice. Please type 'merge' or 'separate'.")

    # === Step 6: Download option ===
    download_choice = input("\nDo you want to download the Excel file? (yes/no): ").strip().lower()

    if download_choice == "yes":
        files.download(output_file)
        print("✅ File downloaded!")
    else:
        print("❌ Download skipped.")
```

# ⚠️ But note the limitations:

- It only extracts tables. If your Word file has paragraphs, titles, or sections outside of tables, they will be ignored.

- If a document is purely text (no tables), the result will be an empty Excel.
