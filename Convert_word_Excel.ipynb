# Script Converter

- Extracts all tables from a Word document.
- Merges tables into a single Excel sheet

```python
import pandas as pd
from docx import Document

# Replace with your uploaded file name
docx_file = list(uploaded.keys())[0]

doc = Document(docx_file)

all_tables = []
for table in doc.tables:
    data = []
    for row in table.rows:
        data.append([cell.text.strip() for cell in row.cells])
    df = pd.DataFrame(data)
    all_tables.append(df)

# Merge all tables into one sheet
merged_df = pd.concat(all_tables, ignore_index=True)

# Save to Excel
output_file = "Word_to_Excel.xlsx"
merged_df.to_excel(output_file, index=False)

```

- Extracts all tables from a Word document.
- Spread tables into a diferent Excel sheet.

```python
import pandas as pd
from docx import Document

# Get uploaded file name
docx_file = list(uploaded.keys())[0]

# Load Word document
doc = Document(docx_file)

# Prepare Excel writer
output_file = "Word_to_Excel_SeparateSheets.xlsx"
writer = pd.ExcelWriter(output_file, engine="openpyxl")

# Loop through tables and save each to its own sheet
for i, table in enumerate(doc.tables, start=1):
    data = []
    for row in table.rows:
        data.append([cell.text.strip() for cell in row.cells])

    df = pd.DataFrame(data)
    sheet_name = f"Table_{i}"
    df.to_excel(writer, index=False, sheet_name=sheet_name)

# Save Excel file
writer.close()

# Download the file
files.download(output_file)
```

# COMBINE SCRIPT


```python
import pandas as pd
from docx import Document
from google.colab import files

# === Step 1: Upload Word file from user device ===
print("📂 Please upload your Word (.docx) file")
uploaded = files.upload()
docx_file = list(uploaded.keys())[0]

# === Step 2: Load Word document ===
doc = Document(docx_file)

# === Step 3: Extract tables into DataFrames ===
all_tables = []
for i, table in enumerate(doc.tables, start=1):
    data = []
    for row in table.rows:
        data.append([cell.text.strip() for cell in row.cells])
    df = pd.DataFrame(data)
    all_tables.append((f"Table_{i}", df))  # keep table name + dataframe

# === Step 4: Ask user preference ===
choice = input("Do you want all tables merged into ONE sheet (type 'merge') or each table in a SEPARATE sheet (type 'separate')? ").strip().lower()

if choice == "merge":
    # Merge all into one sheet
    merged_df = pd.concat([df for _, df in all_tables], ignore_index=True)
    print("\n🔎 Preview of merged data:")
    display(merged_df)   # Colab shows this nicely

    output_file = "Word_to_Excel_Merged.xlsx"
    merged_df.to_excel(output_file, index=False)

elif choice == "separate":
    # Each table goes to its own sheet
    output_file = "Word_to_Excel_SeparateSheets.xlsx"
    writer = pd.ExcelWriter(output_file, engine="openpyxl")

    for sheet_name, df in all_tables:
        df.to_excel(writer, index=False, sheet_name=sheet_name)

    writer.close()

    # Show preview of first sheet
    print(f"\n🔎 Preview of {all_tables[0][0]}:")
    display(all_tables[0][1])

else:
    raise ValueError("Invalid choice. Please type 'merge' or 'separate'.")

# === Step 5: Ask if user wants to download ===
download_choice = input("\nDo you want to download the Excel file? (yes/no): ").strip().lower()
if download_choice == "yes":
    files.download(output_file)
    print("✅ File downloaded!")
else:
    print("❌ Download skipped.")
```

# ⚠️ But note the limitations:

- It only extracts tables. If your Word file has paragraphs, titles, or sections outside of tables, they will be ignored.

- If a document is purely text (no tables), the result will be an empty Excel.
